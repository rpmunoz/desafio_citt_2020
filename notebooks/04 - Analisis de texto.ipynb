# Analizando textos con Azure Text Analytics

## 1. Crear el servicio en Azure

```python
import requests
from pprint import pprint
import os
```

```python
subscription_key = ""
endpoint = ""
```

## 2. Detectar el idioma de un texto

```python
url_lenguaje = endpoint + "/text/analytics/v3.0/languages"
```

```python
texto_espanol = {"id": "1", "text": ""}
```

```python
texto_ingles = {"id": "2", "text": ""}
```

```python
texto_otro_idioma = {"id": "3", "text": ""}
```

```python
documentos = {"documents": [
    texto_espanol,
    texto_ingles,
    texto_otro_idioma
]}
```

```python
cabecera = {"Ocp-Apim-Subscription-Key": subscription_key}
respuesta = requests.post(url_lenguaje, headers=cabecera, json=documentos)
```

```python
lenguajes_detectados = respuesta.json()
pprint(lenguajes_detectados)
```
